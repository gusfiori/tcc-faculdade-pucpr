```python
# Carregamento do dataset

dataset1 = "/content/drive/MyDrive/TCC/D1"
dataset1Train = "/content/drive/MyDrive/TCC/D1/train"
csv_path = "/content/drive/MyDrive/TCC/D1/train_gender.csv"

# Leitura do CSV
df = pd.read_csv(csv_path)

# Converte IDs numéricos para string com 6 dígitos + extensão .png
df['imageId'] = df['imageId'].apply(lambda x: f"{int(x):06d}.png")
```

```python
# Hold-Out
train_df, test_df = train_test_split(
    df,
    train_size=2500,
    test_size=1200,
    stratify=df['gender'],
    random_state=12
)

# Verifica os tamanhos
print(f"Tamanho do conjunto de treino (train_df): {len(train_df)}")
print(f"Tamanho do conjunto de teste (test_df): {len(test_df)}")
```

```python
# converte para grayscale, normaliza para [0,1] e redimensiona para 224×224 (grayscale)

def CarregarEPreprocessar(filename):
    img_path = tf.strings.join([train_dir, filename], separator='/')
    image_bytes = tf.io.read_file(img_path)
    image = tf.image.decode_png(image_bytes, channels=1)
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize_with_pad(image, IMG_SIZE, IMG_SIZE) if MANTER_PROPORCAO else tf.image.resize(image, [IMG_SIZE, IMG_SIZE])
    return image, image

train_dir_str = dataset1Train
train_dir = tf.constant(train_dir_str)

IMG_SIZE = 224
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE
MANTER_PROPORCAO = False
```

```python
# Cria o dataset de treino e teste a partir dos DataFrames
train_dataset = (
    tf.data.Dataset.from_tensor_slices(train_df['imageId'].astype(str).values)
    .map(CarregarEPreprocessar, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)

test_dataset = (
    tf.data.Dataset.from_tensor_slices(test_df['imageId'].astype(str).values)
    .map(CarregarEPreprocessar, num_parallel_calls=AUTOTUNE)
    .batch(BATCH_SIZE)
    .prefetch(AUTOTUNE)
)
```

### Autoencoder 1

```python
autoencoder = models.Sequential(name="Autoencoder")

# ENCODER
autoencoder.add(layers.Input(shape=(224, 224, 1)))
autoencoder.add(layers.Conv2D(32, (3, 3), activation='relu', padding='same'))
autoencoder.add(layers.MaxPooling2D((2, 2), padding='same'))
autoencoder.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
autoencoder.add(layers.MaxPooling2D((2, 2), padding='same'))
autoencoder.add(layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
autoencoder.add(layers.MaxPooling2D((2, 2), padding='same'))

# DECODER
autoencoder.add(layers.Conv2DTranspose(128, (3, 3), strides=2, activation='relu', padding='same'))
autoencoder.add(layers.Conv2DTranspose(64, (3, 3), strides=2, activation='relu', padding='same'))
autoencoder.add(layers.Conv2DTranspose(32, (3, 3), strides=2, activation='relu', padding='same'))
autoencoder.add(layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same'))

autoencoder.compile(optimizer=optimizers.Adam(1e-3), loss='mse')
autoencoder.summary()
```

```python
#Salvar os logs do modelo
log_dir = "/content/drive/MyDrive/TCC/logs"
os.makedirs(log_dir, exist_ok=True)

log_path = os.path.join(log_dir, "treino_autoencoder1.csv")

# Cria o logger
csv_logger = CSVLogger(log_path, append=False)
```

```python
history = autoencoder.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=30,
    callbacks=[csv_logger]
)
```

```python
# Cria o encoder "recortando" as 6 primeiras camadas do autoencoder
encoder_output = autoencoder.layers[6].output
encoder = Model(inputs=autoencoder.layers[0].input, outputs=encoder_output, name="Encoder")
encoder.summary()
```

```python
#Cria o feature_model (encoder + GlobalAveragePooling2D → vetor 128D)
gap = layers.GlobalAveragePooling2D(name="gap")(encoder.output)
feature_model = Model(inputs=encoder.input, outputs=gap, name="FeatureModel")
feature_model.summary()
```

```python
BASE_MODELS = "/content/drive/MyDrive/TCC/models"

# Função para salvar os modelos
def SalvarModelos(encoder, nome_base, pasta=BASE_MODELS, feature_model=None, autoencoder=None):
    os.makedirs(pasta, exist_ok=True)
    caminhos = {}

    caminho_encoder = os.path.join(pasta, f"{nome_base}_encoder.keras")
    encoder.save(caminho_encoder)
    caminhos["encoder"] = caminho_encoder

    if feature_model is not None:
        caminho_feature = os.path.join(pasta, f"{nome_base}_feature.keras")
        feature_model.save(caminho_feature)
        caminhos["feature_model"] = caminho_feature

    if autoencoder is not None:
        caminho_auto = os.path.join(pasta, f"{nome_base}_autoencoder.keras")
        autoencoder.save(caminho_auto)
        caminhos["autoencoder"] = caminho_auto

    return caminhos
```

```python
# Amostrar estratificado preservando índices (2000 train, 1000 test, 500 val)
rng = np.random.RandomState(12)

def amostrar_estratificado_keep_index(df_src, n_total, label_col="gender"):
    partes = []
    for cls, g in df_src.groupby(label_col):
        prop = len(g) / len(df_src)
        n_cls = max(1, int(round(prop * n_total)))
        partes.append(g.sample(n=min(n_cls, len(g)), random_state=rng.randint(0, 10**6)))
    out = pd.concat(partes).sample(frac=1.0, random_state=12)  # não reseta índice!
    if len(out) > n_total:
        out = out.iloc[:n_total]
    return out

train_df_2000 = amostrar_estratificado_keep_index(train_df, n_total=2000, label_col="gender")
test_df_1000  = amostrar_estratificado_keep_index(test_df,  n_total=1000,  label_col="gender")
resto_train = train_df.drop(train_df_2000.index)
val_df_500  = amostrar_estratificado_keep_index(resto_train, n_total=500, label_col="gender")

print("Tamanhos (reduzidos):", len(train_df_2000), len(test_df_1000), len(val_df_500))
```

```python
#Datasets só de imagem (D1 reduzido)
AUTOTUNE = tf.data.AUTOTUNE
BATCH_SIZE = 32

def DatasetSoImagens(ids_series):
    ds = tf.data.Dataset.from_tensor_slices(ids_series.astype(str).values)
    ds = ds.map(CarregarEPreprocessar, num_parallel_calls=AUTOTUNE).map(lambda x, y: x)
    return ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

train_images_ds = DatasetSoImagens(train_df_2000['imageId'])
test_images_ds  = DatasetSoImagens(test_df_1000['imageId'])
val_images_ds = DatasetSoImagens(val_df_500['imageId'])
```

```python
def PlotarMatrizConfusao(y_true, y_pred, labels=None, normalizado=False, titulo="Matriz de Confusão"):
    if labels is None:
        labels = np.unique(np.concatenate([np.asarray(y_true), np.asarray(y_pred)]))

    cm = confusion_matrix(
        y_true, y_pred, labels=labels,
        normalize='true' if normalizado else None
    )

    fig, ax = plt.subplots(figsize=(5, 4))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(ax=ax, colorbar=False, values_format=".2f" if normalizado else "d")
    ax.set_title(titulo + (" (normalizada)" if normalizado else ""))
    plt.tight_layout()
    plt.show()
```

```python
# SVM
def FormatarSaida(x: float) -> str:
    return f"{x*100:.2f}%".replace('.', ',')

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_val_std   = scaler.transform(X_val)
X_test_std  = scaler.transform(X_test)

# Treina SVM no treino
svm = SVC(kernel='linear', random_state=12)
svm.fit(X_train_std, y_train)

y_va_pred = svm.predict(X_val_std)
acc_va = accuracy_score(y_val, y_va_pred)
f1_va  = f1_score(y_val, y_va_pred, average='weighted')
print("Validação")
print("Acurácia (val):", FormatarSaida(acc_va))
print("F1  (val):",    FormatarSaida(f1_va))

y_te_pred = svm.predict(X_test_std)
acc_te = accuracy_score(y_test, y_te_pred)
f1_te  = f1_score(y_test, y_te_pred, average='weighted')
print("\nTeste")
print("Acurácia (test):", FormatarSaida(acc_te))
print("F1  (test):",     FormatarSaida(f1_te))
```

```python
def split_70_15_15_por_classe(df, label_col, rnd=12):
    rng = np.random.RandomState(rnd)
    partes_train, partes_val, partes_test = [], [], []
    for cls, g in df.groupby(label_col):
        n = len(g)
        if n == 1:
            g_shuf = g.sample(frac=1, random_state=rng.randint(0, 10**6))
            partes_train.append(g_shuf); continue
        n_val  = max(1, int(round(0.15 * n)))
        n_test = max(1, int(round(0.15 * n)))
        g_shuf = g.sample(frac=1, random_state=rng.randint(0, 10**6))
        partes_val.append(g_shuf.iloc[:n_val])
        partes_test.append(g_shuf.iloc[n_val:n_val+n_test])
        partes_train.append(g_shuf.iloc[n_val+n_test:])
    df_train = pd.concat(partes_train).sample(frac=1, random_state=rng.randint(0, 10**6)).reset_index(drop=True)
    df_val   = pd.concat(partes_val).sample(frac=1, random_state=rng.randint(0, 10**6)).reset_index(drop=True)
    df_test  = pd.concat(partes_test).sample(frac=1, random_state=rng.randint(0, 10**6)).reset_index(drop=True)
    return df_train, df_val, df_test

df2_train, df2_val, df2_test = split_70_15_15_por_classe(df2, label_col="gender", rnd=12)
print("Split D2 ->", len(df2_train), len(df2_val), len(df2_test))
```

```python
# Carrega e extrai um dataset do Kaggle (robusto)
import os, json, subprocess, sys
from pathlib import Path

def carregarDatasetKaggle(kaggle_dataset, zip_name, data_root, data_dir, kaggle_json_path=None):
    """
    Baixa e extrai um dataset do Kaggle em data_root e verifica data_dir.
    Se kaggle_json_path for fornecido, usa-o para configurar ~/.kaggle/kaggle.json.
    Caso contrário, tenta usar as variáveis de ambiente KAGGLE_USERNAME e KAGGLE_KEY.
    """
    data_root = Path(data_root)
    data_dir  = Path(data_dir)
    work_dir  = Path("/content")  # opcional; apenas para logging
    print("Work dir:", work_dir)
    print("Data root:", data_root)
    print("Data dir esperado:", data_dir)

    # 1) Garante pacote kaggle
    try:
        import kaggle  # noqa
    except Exception:
        print("Instalando pacote kaggle...")
        subprocess.run([sys.executable, "-m", "pip", "install", "-q", "kaggle"], check=True)

    # 2) Configura credencial kaggle.json
    kaggle_dir = Path.home() / ".kaggle"
    kaggle_dir.mkdir(parents=True, exist_ok=True)
    cred_file = kaggle_dir / "kaggle.json"

    if kaggle_json_path:
        kaggle_json_path = Path(kaggle_json_path)
        if not kaggle_json_path.exists():
            raise FileNotFoundError(f"kaggle.json não encontrado em: {kaggle_json_path}")
        cred_file.write_bytes(kaggle_json_path.read_bytes())
        print("kaggle.json copiado de:", kaggle_json_path)
    else:
        # tenta variáveis de ambiente
        u = os.environ.get("KAGGLE_USERNAME")
        k = os.environ.get("KAGGLE_KEY")
        if u and k:
            cred_file.write_text(json.dumps({"username": u, "key": k}))
            print("kaggle.json criado a partir de variáveis de ambiente.")
        elif not cred_file.exists():
            raise RuntimeError(
                "Credenciais do Kaggle ausentes. Forneça kaggle_json_path ou exporte "
                "KAGGLE_USERNAME e KAGGLE_KEY."
            )

    os.chmod(cred_file, 0o600)

    # 3) Baixa o zip se necessário
    data_root.mkdir(parents=True, exist_ok=True)
    zip_path = data_root / zip_name

    if data_dir.exists() and any(data_dir.iterdir()):
        print("Dataset já parece extraído em", data_dir)
        return

    if not zip_path.exists():
        print(f"Baixando dataset {kaggle_dataset} do Kaggle...")
        cmd = ["kaggle", "datasets", "download", "-d", kaggle_dataset, "-p", str(data_root), "--force"]
        proc = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
        print(proc.stdout)
        if proc.returncode != 0:
            raise RuntimeError("Falha no download via Kaggle. Verifique credenciais e dataset id.")

    # 4) Descompacta
    if not data_dir.exists() or not any(data_dir.iterdir()):
        print("Descompactando...")
        proc = subprocess.run(["unzip", "-q", str(zip_path), "-d", str(data_root)])
        if proc.returncode != 0:
            raise RuntimeError(f"Falha ao descompactar: {zip_path}")
        print("Extração concluída.")

    # 5) Verifica conteúdo final
    if not data_dir.exists() or not any(data_dir.iterdir()):
        raise RuntimeError(f"Após extração, {data_dir} ainda está vazio. Confira estrutura do zip.")
    print("OK! Dados prontos em:", data_dir)
```

```python
# Lê CSV do NIH e retorna DataFrame com ["imageId", "gender"]
def LerCSVNIHGenero(csv_path: str, img_dir: str):
    df = pd.read_csv(csv_path)

    # Descobre a coluna de gênero
    possiveis_cols = ["gender", "Gender", "Patient Gender", "patient_gender", "Patient Gender "]
    col_gender = next((c for c in possiveis_cols if c in df.columns), None)
    if col_gender is None:
        raise ValueError(f"Não encontrei coluna de gênero. Colunas disponíveis: {list(df.columns)}")

    # Normaliza rótulos de gênero
    def NormalizarGenero(x):
        s = str(x).strip().lower()
        if s in {"m", "male"}: return "male"
        if s in {"f", "female"}: return "female"
        return s

    # Padroniza nome do arquivo de imagem
    if "Image Index" in df.columns:
        df = df.rename(columns={"Image Index": "imageId"})
    elif "imageId" not in df.columns:
        raise ValueError("Não encontrei coluna de arquivo de imagem ('Image Index' ou 'imageId').")

    df["imageId"] = df["imageId"].astype(str)
    df["gender"]  = df[col_gender].apply(NormalizarGenero)

    # Mantém só imagens que existem na pasta
    df["__exists__"] = df["imageId"].apply(lambda f: os.path.exists(os.path.join(img_dir, f)))
    df = df[df["__exists__"]].drop(columns="__exists__").reset_index(drop=True)

    return df[["imageId", "gender"]]
```

```python
# Split 70/15/15 estratificado por 'gender' (robusto para classes raras)
def Split_70_15_15Genero(df: pd.DataFrame, seed: int = 12):
    vc = df["gender"].value_counts()

    # Caso tenha classe com <2 amostras ou apenas 1 classe → split manual por classe
    if (vc < 2).any() or len(vc) < 2:
        rng = np.random.RandomState(seed)
        partes_train, partes_val, partes_test = [], [], []

        for _, grp in df.groupby("gender"):
            n = len(grp)
            if n == 1:
                partes_train.append(grp)
                continue

            n_val  = max(1, int(round(0.15 * n)))
            n_test = max(1, int(round(0.15 * n)))
            if n_val + n_test >= n:  # evita estouro
                n_val, n_test = 1, 1
            n_train = n - n_val - n_test

            grp = grp.sample(frac=1.0, random_state=rng.randint(0, 10**6)).reset_index(drop=True)
            partes_val.append(grp.iloc[:n_val])
            partes_test.append(grp.iloc[n_val:n_val+n_test])
            partes_train.append(grp.iloc[n_val+n_test:])

        df_train = pd.concat(partes_train, ignore_index=True).sample(frac=1.0, random_state=seed).reset_index(drop=True)
        df_val   = pd.concat(partes_val,   ignore_index=True).sample(frac=1.0, random_state=seed).reset_index(drop=True)
        df_test  = pd.concat(partes_test,  ignore_index=True).sample(frac=1.0, random_state=seed).reset_index(drop=True)
        return df_train, df_val, df_test

    # Caminho normal (todas as classes têm >= 2 amostras)
    df_train, df_tmp = train_test_split(
        df, test_size=0.30, stratify=df["gender"], random_state=seed
    )
    df_val, df_test = train_test_split(
        df_tmp, test_size=0.50, stratify=df_tmp["gender"], random_state=seed
    )
    return (df_train.reset_index(drop=True),
            df_val.reset_index(drop=True),
            df_test.reset_index(drop=True))
```

```python
# Amostragem fixa estratificada (mantém proporção de 'gender')
def amostrar_fix_estratificado(df_src, n_total, label_col="gender", seed=12):
    rng = np.random.RandomState(seed)
    df_src = df_src.copy()
    df_src["__idx__"] = np.arange(len(df_src))
    partes = []
    for cls, g in df_src.groupby(label_col):
        prop = len(g) / len(df_src)
        n_cls = max(1, int(round(prop * n_total)))
        partes.append(g.sample(n=min(n_cls, len(g)), random_state=rng.randint(0, 10**6)))
    sub = pd.concat(partes).sample(frac=1.0, random_state=seed)
    if len(sub) > n_total:
        sub = sub.iloc[:n_total]
    return sub, sub["__idx__"].values

df3_train_2000, idx_sub_train = amostrar_fix_estratificado(df3_train, n_total=2000)
df3_test_1000,  idx_sub_test  = amostrar_fix_estratificado(df3_test,  n_total=1000)
print(len(df3_train_2000), len(df3_test_1000))
```

```python
def plot_reconstrucoes_from_df(
    ae_path: str,
    df_imgs,              # DataFrame com coluna 'imageId'
    img_dir: str,         # pasta das imagens
    n: int = 3,
    seed: int = 42,
    img_size=(224, 224),
    channels: int = 1,    # 1 se AE grayscale, 3 se AE RGB
    manter_proporcao: bool = True,
):
    # 1) Carrega AE completo
    ae = load_model(ae_path, compile=False)

    # 2) Amostra n imagens do DF
    df_subset = df_imgs.sample(n=min(n, len(df_imgs)), random_state=seed)
    ids = df_subset["imageId"].astype(str).values

    AUTOTUNE = tf.data.AUTOTUNE
    img_dir_const = tf.constant(str(img_dir))

    def _carregar_preprocessar(img_id):
        img_path = tf.strings.join([img_dir_const, img_id], separator=os.sep)
        raw = tf.io.read_file(img_path)
        img = tf.image.decode_image(raw, channels=channels, expand_animations=False)
        img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
        if manter_proporcao:
            img = tf.image.resize_with_pad(img, img_size[0], img_size[1])
        else:
            img = tf.image.resize(img, img_size)
        return img

    ds = (tf.data.Dataset.from_tensor_slices(ids)
          .map(_carregar_preprocessar, num_parallel_calls=AUTOTUNE)
          .batch(len(ids))
          .prefetch(AUTOTUNE))

    orig = next(iter(ds)).numpy()              # (n,H,W,C) [0,1]
    recon = ae.predict(orig, verbose=0)        # (n,H,W,C)
    orig  = np.clip(orig,  0.0, 1.0)
    recon = np.clip(recon, 0.0, 1.0)

    # 3) Métricas
    ssim_vals = tf.image.ssim(orig, recon, max_val=1.0).numpy()
    psnr_vals = tf.image.psnr(orig, recon, max_val=1.0).numpy()

    # 4) Plot robusto (sem título)
    n_plot = orig.shape[0]
    is_gray = (orig.shape[-1] == 1)

    fig, axes = plt.subplots(2, n_plot, figsize=(1.9*n_plot, 5.2), constrained_layout=False)
    if n_plot == 1:
        axes = np.array([[axes[0]], [axes[1]]])

    for i in range(n_plot):
        ax1 = axes[0, i]
        ax1.imshow(orig[i].squeeze(), cmap="gray", vmin=0, vmax=1) if is_gray else ax1.imshow(orig[i])
        ax1.set_title("Original", fontsize=11)
        ax1.set_xticks([]); ax1.set_yticks([])

        ax2 = axes[1, i]
        ax2.imshow(recon[i].squeeze(), cmap="gray", vmin=0, vmax=1) if is_gray else ax2.imshow(recon[i])
        ax2.set_title("Reconstruída", fontsize=11)
        ax2.set_xticks([]); ax2.set_yticks([])
        # SSIM na 1ª linha, PSNR na 2ª
        ax2.set_xlabel(f"SSIM={ssim_vals[i]:.4f}\nPSNR={psnr_vals[i]:.2f} dB",
                       fontsize=10, labelpad=6)

    # dá espaço vertical pros xlabels em 2 linhas
    plt.subplots_adjust(top=0.96, hspace=0.55, bottom=0.14)
    plt.show()
```
